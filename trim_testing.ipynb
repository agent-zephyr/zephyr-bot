```python
import subprocess
# ffmpeg -ss 30 -i input.wmv -c copy -t 10 output.wmv
command = [
    "ffmpeg",
    "-ss", "30",          # Start time at 30 seconds
    "-i", "input.mp4",    # Input file
    "-c", "copy",         # Copy codec
    "-t", "10",           # Duration of 10 seconds
    "output.mp4"          # Output file
]

try:
    subprocess.run(command, check=True)
    command[2] = "40"
    command[8] = "11"
    command[9] = "output2.mp4"
    subprocess.run(command, check=True)
    print("FFmpeg command executed successfully.")
except subprocess.CalledProcessError as e:
    print(f"Error occurred: {e}")
```

```python
command = [
    "ffmpeg",
    "-f", "concat",       # Use concat demuxer
    "-safe", "0",         # Allow potentially unsafe file paths
    "-i", "file_list.txt",# Input file list
    "-c:v", "libx264",    # Re-encode during concatenation
    "-crf", "23",
    "-preset", "fast",
    "-c:a", "aac",
    "-b:a", "128k",
    "final.mp4"
]

# Run the command
try:
    subprocess.run(command, check=True)
    print("FFmpeg command executed successfully.")
except subprocess.CalledProcessError as e:
    print(f"Error occurred: {e}")
```

```python
def cut_segment(input_file, start_time, duration, output_file):
    command = [
        "ffmpeg",
        "-ss", str(start_time),  # Start time
        "-i", input_file,        # Input file
        "-t", str(duration),     # Duration
        "-c", "copy",            # Copy codec without re-encoding
        "-reset_timestamps", "1",# Reset timestamps
        output_file              # Output file
    ]
    subprocess.run(command, check=True)

cut_segment("input.mp4", 30, 10, "output.mp4")
cut_segment("input.mp4", 40, 11, "output2.mp4")

command = [
    "ffmpeg",
    "-f", "concat",       # Use concat demuxer
    "-safe", "0",         # Allow potentially unsafe paths
    "-i", "file_list.txt",# Input file list
    "-c", "copy",         # Copy streams without re-encoding
    "final.mp4"           # Output file
]

try:
    subprocess.run(command, check=True)
    print("Video segments concatenated successfully into 'final.mp4'.")
except subprocess.CalledProcessError as e:
    print(f"Error during concatenation: {e}")
```

```python
def trim_and_reencode(input_file, start_time, duration, output_file):
    command = [
        "ffmpeg",
        "-ss", str(start_time),  # Start time
        "-i", input_file,        # Input file
        "-t", str(duration),     # Duration
        "-c:v", "libx264",       # Re-encode video to H.264
        "-crf", "23",            # Quality factor for x264
        "-preset", "fast",       # Encoding speed/quality tradeoff
        "-c:a", "aac",           # Re-encode audio
        "-b:a", "128k",          # Audio bitrate
        "-keyint_min", "30",     # Minimum keyframe interval
        "-g", "60",              # Maximum keyframe interval (frames)
        "-sc_threshold", "0",    # Disable scene change detection for consistent keyframes
        output_file
    ]
    subprocess.run(command, check=True)


trim_and_reencode("input.mp4", 30, 10, "output.mp4")
trim_and_reencode("input.mp4", 40, 10, "output2.mp4")
```
